# QUES 1 

```python
# Here we are converting the date columns to datetime objects

df['Time'] = pd.to_datetime(df['Time'])
df['Time Out'] = pd.to_datetime(df['Time Out'])
```

```python
# Creating a new column to represent the date of each entry (ignoring the time)
df['Date'] = df['Time'].dt.date
```

```python
def consecutive_days_worked(employee_data):
    consecutive_days = 0
    prev_date = None

    for date in employee_data['Date']:
        if prev_date is None:
            consecutive_days = 1
        elif (date - prev_date).days == 1:
            consecutive_days += 1
        else:
            consecutive_days = 1
        
        if consecutive_days == 7:
            return True
        
        prev_date = date

    return False
```

```python
unique_employees = df['Employee Name'].unique()

for employee in unique_employees:
    employee_data = df[df['Employee Name'] == employee]
    if consecutive_days_worked(employee_data):
        print(f"Employee: {employee}, Position: {employee_data['Position ID'].values[0]}")
    else:
        print(f"No 7 consecutive days worked for Employee: {employee}")
        
# there are no employees who has worked for 7 consecutive days in our dataset.
```

# QUES 2

```python
# Sorting the DataFrame by Employee Name and Time
df.sort_values(['Employee Name', 'Time'], inplace=True)

# Calculating the time between shifts for each employee
df['Time Between Shifts'] = df.groupby('Employee Name')['Time'].diff()

# Filtering the DataFrame to include only rows with time between shifts less than 10 hours but greater than 1 hour
filtered_df = df[(df['Time Between Shifts'] < pd.Timedelta(hours=10)) & (df['Time Between Shifts'] > pd.Timedelta(hours=1))]

# Getting the unique employees who meets the criteria
unique_employees = filtered_df['Employee Name'].unique()

for employee in unique_employees:
    employee_data = filtered_df[filtered_df['Employee Name'] == employee]
    print(f"Employee: {employee}, Position: {employee_data['Position ID'].values[0]}")
```

# QUES 3

```python
# Calculating the duration of each shift in hours
df['Shift Duration'] = (df['Time Out'] - df['Time']).dt.total_seconds() / 3600

# Filtering the DataFrame to include only rows where the shift duration is greater than 14 hours
filtered_df = df[df['Shift Duration'] > 14]

# Getting the unique employees who meets the criteria
unique_employees = filtered_df['Employee Name'].unique()

for employee in unique_employees:
    employee_data = filtered_df[filtered_df['Employee Name'] == employee]
    print(f"Employee: {employee}, Position: {employee_data['Position ID'].values[0]}, Shift Duration: {employee_data['Shift Duration'].values[0]:.2f} hours")
```
